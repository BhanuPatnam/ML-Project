```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def convert_sqfeet_tonum(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4=df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(convert_sqfeet_tonum)
df4.head(40)
```
